# Data Set
The data set contains 150 observations of iris flowers.There are four columns of measurements of the flowers in cm.The fifth column is the species of the flower observed.All observed flowers belong to one of three species.

# Seaborn import for plotting Linear Regression.

```python
sns.lmplot(x='SepalLength',y='SepalWidth',data=iris)
```

```python
sns.lmplot(x='PetalLength',y='PetalWidth',data=iris)
```

# Generate a residual plot

```python
sns.residplot(x=iris['SepalLength'],y=iris['SepalWidth'],color='red')
```

```python
sns.residplot(x=iris['PetalLength'],y=iris['PetalWidth'],color='b')
```

```python
iris.plot(kind='density',subplots=True,layout=(3,3),sharex=False)
plt.show()
```

# Drow regression plot

```python
sns.regplot(x=iris['SepalLength'],y=iris['SepalWidth'],data='iris',color='red',label='Order_1',order=1)
```

```python
sns.regplot(x=iris['PetalLength'],y=iris['PetalWidth'],data='iris',color='green',label='Order_1',order=1)
```

```python
sns.regplot(x=iris['SepalLength'],y=iris['SepalWidth'],data='iris',color='blue',label='Order 1',order=1)
sns.regplot(x=iris['SepalLength'],y=iris['SepalWidth'],data='iris',color='yellow',label='Order 2',order=2)
sns.regplot(x=iris['SepalLength'],y=iris['SepalWidth'],data='iris',color='k',label='Order 3',order=3)
```

# plot a linear regression betwwen the variables of iris dataset by specifing the hue. 

```python
sns.lmplot(x='SepalLength',y='SepalWidth',data=iris, hue='Class', palette='Set1')
```

# We plot a linear regression between the variables of iris dataset grouped by row-wise.

```python
sns.lmplot(x='SepalLength',y='SepalWidth',data=iris,row='Class')
```

# We make a striplot of SepalLength,SepalWidth,PetalLength and PetalWidth grouped by Class.

```python
plt.subplot(2,2,1)
sns.stripplot(x='Class',y='SepalLength',data=iris)

plt.subplot(2,2,2)
sns.stripplot(x='Class',y='SepalWidth',data=iris)
```

```python
plt.subplot(2,2,2)
sns.stripplot(x='Class',y='SepalWidth',data=iris,jitter=True,size=2)
```

# generate a swarmplot

```python
sns.swarmplot(x='Class',y='SepalWidth',data=iris)
sns.swarmplot(x='Class',y='SepalLength',data=iris)
```
